```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import mpld3

%matplotlib inline
```

```python
# Loading in 2021-2022 Regular Season Games

df = pd.read_csv(r"C:\Users\hunte\PythonDAProjects\NBAPlayerStats\22RegularNBAPlayerStats.csv", encoding = 'ISO-8859-1', delimiter = ';')

df.head(10)
```

```python
# Expanding out the columns of the data frame to see all
pd.set_option('display.max_columns', None)

df.head(10)
```

```python
# Seeing if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing * 100)))
```

```python
# Showing the data types of our variabels
print(df.dtypes)
```

```python
# Checking to see if our data frame has any duplicates
duplicates = df.duplicated().sum()

print(duplicates)
```

```python
# Sorting data frame based on highest average scorer
df.sort_values(by = ['PTS'], inplace = False, ascending = False).head(20)
```

```python
# Sorting data frame based on highest average 3 point percentage
df.sort_values(by = ['3P%'], inplace = False, ascending = False).head(20)
```

```python
# I want the data frame to be sorted the same way but include only high volume and high scoring players
mask = (df['3PA'] > 5) & (df['PTS'] > 20)

highVol3_df = df[mask]
highVol3_df = highVol3_df.sort_values(by=['3P%'], inplace=False, ascending=False).head(20)

highVol3_df.head(10)
```

```python
# Lastly I don't want to include players multiple times which have been on seperarte teams

uniqueHighVol_df = highVol3_df.drop_duplicates(subset = ['Player'], keep = 'first')

uniqueHighVol_df.sort_values(by = ['3P%'], ascending = False).head(10)
```

# Creating a Data Frame Without Duplicates

```python
# I'm deciding to keep the players statistics with the highest points from players who have played on multiple teams in this season
noDuplicates_df = df.copy()
noDuplicates_df = noDuplicates_df.sort_values(by=['PTS'], inplace=False, ascending=False)
noDuplicates_df = noDuplicates_df.drop_duplicates(subset=['Player'], keep='first')

noDuplicates_df.head(10)
```

# Interactive Scatter Plot Between Average Points, Assists, and Rebounds per Game

```python
# Creating figure and axis
fig, ax = plt.subplots(figsize=(10, 8))

# Scatter Plot
plot = sns.scatterplot(x = 'PTS', y = 'AST', size = 'TRB', hue = 'Tm', style = 'Tm', data = noDuplicates_df,  alpha = 0.7, legend = 'brief', palette = 'bright', edgecolor = 'black', linewidth = 0.5, sizes = (50, 500), ax = ax)

# Adding a tool tip which allows for the name of the player and their team to be displayed
if plot:
    tooltip = mpld3.plugins.PointHTMLTooltip(plot.get_children()[0],
                                              labels=[f'<div>{p}</div><div>{t}</div>' for p, t in zip(noDuplicates_df['Player'], noDuplicates_df['Tm'])])
    mpld3.plugins.connect(plt.gcf(), tooltip)

# Legend for the plot
plt.legend(ncol = 4, title = "Teams With Icon Size Relating to Total Rebounds")

# Displaying graph
mpld3.display()
```

This graph shows each indivdual players average statistics in Points, Assists, and Rebounds. The Points and Assists can be found on the x and y axis, while rebounds correlates to how large the point is on the graph. To represent the team each player is on, there is a color and shape combination for each team. If you hover over a point the players name will display along with their team.

# Box Plot of Player Minutes by Position

```python
# I'm choosing to filter out the players which identify with more than one position
pfilter = ['PG', 'SG', 'C', 'PF', 'SF']
position_df = df[df['Pos'].isin(pfilter)]

# Sorting values by position
position_df = position_df.sort_values('Pos', key=lambda x: x.map({k: i for i, k in enumerate(pfilter)}))

# Creating box plot
ax = sns.boxplot(x = 'Pos', y = 'MP', data = position_df, showfliers = False, medianprops = {'color': 'black'})

# Set the plot title
ax.set(title = 'Minutes Played by Position')

# Takingout line above and to the right
sns.despine()

# Graph labels
plt.xlabel('Positions')

plt.ylabel('Minutes Played')

# Displaying Graph
plt.show()
```
